```python


from google.colab import files
import pandas as pd
import io

uploaded = files.upload()


```

```python
#1. Compute for a mean, median, mode, variance and standard devation

csv = pd.read_csv(io.BytesIO(uploaded['StudentsPerformanceFor1stLabExam (1).csv']))

print("\nThe Mean OF Math Score is : ", csv['math score'].mean())
print("The Median OF Math Score is : ", csv['math score'].median())
print("The Mode OF Math Score is : ", csv['math score'].mode())
print("The Variance OF Math Score is : ", csv['math score'].var())
print("The Standard Devation OF Math Score is : ", csv['math score'].std())

#---------------------------------------------------------------------------------
print("\nThe Mean OF Reading Score is : ", csv['reading score'].mean())
print("The Median OF Reading Scoree is :", csv['reading score'].median())
print("The Mode OF Reading Score is : ", csv['reading score'].mode())
print("The Variance OF Reading Score is : ", csv['reading score'].var())
print("The Standard Devation OF Reading Score is : ", csv['reading score'].std())

#---------------------------------------------------------------------------------

print("\nThe Mean OF Writing Score is : ", csv['writing score'].mean())
print("The Median OF Writing Scoree is : ", csv['writing score'].median())
print("The Mode OF Writing Score is : ", csv['writing score'].mode())
print("The Variance OF Writing Score is : ", csv['writing score'].var())
print("The Standard Devation OF Writing Score is : ", csv['writing score'].std())
```
