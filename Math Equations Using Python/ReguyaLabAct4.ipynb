```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10
p = 1/5
k = 7

x = np.arange(0, n + 1)


probabilities = binom.pmf(x, n, p)

print(probabilities)
plt.plot(x, probabilities, color='blue', marker='x', linestyle='-')
plt.title('Binomial Distribution (n=10, p=1/5)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
```

```python
import numpy as np
from scipy.stats import binom


n = 12
p = 1/6
k = 5


probability = binom.pmf(k, n, p)


print(f"The probability of getting 3 five times in 12 rolls is: {probability:.4f}")
```
