```python
#Activity 4 Probability of rolling a specific number on a fair six-sided dice
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

outcome = [1,2,3,4,5,6]
prob = len(outcome)
total = outcome[0] / prob

plt.bar(prob , total , color = 'red')
plt.title(' Probability of rolling a specific number on a fair six-sided dice ')
plt.xlabel('Die Face')
plt.ylabel('fair six-sided dice')
plt.ylim(0,1)

```

```python
#Activity 5 probability of getting a head when tossing a coin
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



outcomes = [1,2]

total_outcomes = len(outcomes)


ok = outcomes[0]

probability = ok  / total_outcomes

plt.bar(total_outcomes, probability)
plt.title('probability of getting a head when tossing a coin ')
plt.xlabel('Die Face')
plt.ylabel('Probability')
plt.ylim(0,1)
```

```python
# Activity 6 probability of getting a jacking in a deck of 52 cards
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
jacks = 4

cards = 52


total = jacks / cards
plt.bar(cards, total)
plt.title('Gettin a Probability ')
plt.xlabel(' 52 cards')
plt.ylabel('Probability')
plt.ylim(0,1)
```
